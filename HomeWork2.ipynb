```python
import cv2
import numpy as np

# Read the video
cap_red = cv2.VideoCapture ('tracking1.avi')

# Define range of color in HSV
lower_red = np.array ([172, 50, 50])
upper_red = np.array ([180, 255, 255])
    
lower_yellow = np.array ([20, 50, 50])
upper_yellow = np.array ([30, 255, 255])

# Loop for present and tracking the video
while (cap_red.isOpened()) :
    ret, frame_red = cap_red.read()
    # If the video stream ends exit the loop
    if not ret :
        break
    # Convert video color to hsv from rgb
    hsv_red = cv2.cvtColor (frame_red, cv2.COLOR_BGR2HSV)    
    # Masking the red color
    mask_red = cv2.inRange (hsv_red, lower_red, upper_red)    
    # Remain red color use bitwise_and
    res_red = cv2.bitwise_and (frame_red, frame_red, mask = mask_red)    
    # Show frame_red
    cv2.imshow ('frame', frame_red)
    # Show masked frame
    cv2.imshow ('tracking', res_red)
    # Set wait time and, exit when the input key is ESC
    k = cv2.waitKey (42) & 0xFF
    if (k == 27) :
        break

cap_red.release()

cap_yellow = cv2.VideoCapture ('tracking2.avi')
    
while (cap_yellow.isOpened()) :
        ret, frame_yellow = cap_yellow.read()
        if not ret :
            break
        hsv_yellow = cv2.cvtColor (frame_yellow, cv2.COLOR_BGR2HSV)
        mask_yellow = cv2.inRange (hsv_yellow, lower_yellow, upper_yellow)
        res_yellow = cv2.bitwise_and (frame_yellow, frame_yellow, mask = mask_yellow)
        cv2.imshow ('frame', frame_yellow)
        cv2.imshow ('tracking', res_yellow)
        k = cv2.waitKey (42) & 0xFF
        if (k == 27) :
            break        
            
cap_yellow.release()
        
cap_red_bg = cv2.VideoCapture ('tracking3.avi')

while (cap_red_bg.isOpened()) :
    ret, frame_red_bg = cap_red_bg.read()
    if not ret :
        break
    hsv_red_bg = cv2.cvtColor (frame_red_bg, cv2.COLOR_BGR2HSV)
    mask_red_bg = cv2.inRange (hsv_red_bg, lower_red, upper_red)
    res_red_bg = cv2.bitwise_and (frame_red_bg, frame_red_bg, mask = mask_red_bg)
    cv2.imshow ('frame', frame_red_bg)
    cv2.imshow ('tracking', res_red_bg)
    k = cv2.waitKey (42) & 0xFF
    if (k == 27) :
        break
    
cap_red_bg.release()

cap_yellow_bg = cv2.VideoCapture ('tracking4.avi')

while (cap_yellow_bg.isOpened()) :
    ret, frame_yellow_bg = cap_yellow_bg.read()
    if not ret :
        break
    hsv_yellow_bg = cv2.cvtColor (frame_yellow_bg, cv2.COLOR_BGR2HSV)
    mask_yellow_bg = cv2.inRange (hsv_yellow_bg, lower_yellow, upper_yellow)
    res_yellow_bg = cv2.bitwise_and (frame_yellow_bg, frame_yellow_bg, mask = mask_yellow_bg)
    cv2.imshow ('frame', frame_yellow_bg)
    cv2.imshow ('tracking', res_yellow_bg)
    k = cv2.waitKey (42) & 0xFF
    if (k == 27) :
        break
    
cap_yellow_bg.release()        
cv2.destroyAllWindows ()
```
