```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread ('sIMG_2180.jpg', 0)
edges = cv2.Canny(img, 100, 250)

plt.subplot(121), plt.imshow (img, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow (edges, cmap='gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show ()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def nothing (x) :
    pass

img = cv2.imread ('sIMG_2180.jpg', 0)
cv2.namedWindow('edge_th')
img2 = cv2.imread ('sIMG_2180.jpg', 0)
cv2.namedWindow('edge_va')

# Create trackbars for color change
cv2.createTrackbar ('Threshold', 'edge_th', 0, 255, nothing)
cv2.createTrackbar ('Min', 'edge_va', 0, 255, nothing)
cv2.createTrackbar ('Max', 'edge_va', 0, 255, nothing)

while (1) :
    k = cv2.waitKey(1) & 0xFF
    if k == 27 :
        break
        
    # Get current potisions of four trackbars
    th = cv2.getTrackbarPos ('Threshold', 'edge_th')
    mi = cv2.getTrackbarPos ('Min', 'edge_va')
    ma = cv2.getTrackbarPos ('Max', 'edge_va')
    ret, thresh = cv2.threshold (img, th, 255, cv2.THRESH_BINARY)
    edges_va = cv2.Canny(img2, mi, ma)
    edges_th = cv2.Canny(thresh, 100, 200)
    
    cv2.imshow ('edge_va', edges_va)
    cv2.imshow ('edge_th', edges_th)
        
cv2.destroyAllWindows()    
```
