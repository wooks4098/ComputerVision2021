# Inheritance

## Subclasses
<h3>
    <ul>
        <li>A class can extend the definition of another class</li>
        <h4>
            <ul>
                <li>Allows use (or extension) of methods and attributes already defined in the prvious one.</li>
                <li>New class : subclass, Original : parent, ancestor or superclass
            </ul>
        </h4>
        <li>To define a subclass, put the name of the superclass in parentheses after the subclass's name on the first line of the definition.</li>
         <code>
             class ai_student (student) :
        </code>
        <h4>
            <ul>
                <li>Python has no 'extends' keyword like java</li>
                <li>Multiple inheritance is supported</li>
            </ul>
        </h4>
    </ul>
</h3>

<h2>Redefining Methods</h2>
<h3>
    <ul>
        <li>Very similar to over-riding methods in Java</li>
        <li>To redefine a method of the parent class, include a new definition using the same name in the subclass.</li>
        <h4><ul><li>The old code won't get excuted.</li></ul></h4>
        <li>To excute the method in the parent class in addition to new code for some methods, explicitly call the parent's version of the method.</li>
        <h4>
            <code>
                parentClass.methodName (self, a, b, c, d)
            </code>
            <ul>
            <li>The only time you ever explicitly pass 'self' as an argument is when calling a method of an ancestor.</li>
            </ul>
        </h4>
    </ul>
</h3>

<h2>Extending __init__</h2>
<h3>
    <ul>
        <li>Very similar to Java</li>
        <li>Commonly, the ancestor's __init__ method is executed in addition to new commands.</li>
        <li><i>Must be done explicitly</i></li>
        <li>You'll often see something like this in the __init__ method of subclassed :
            <code>
                parentClass.__init__(self, x, y)
            </code><br>
        where parentClass is the name of the parent's class.</li>
    </ul>
</h3>
    
<h2>Private Data and Methods</h2>
<h3>
   <ul>
       <li> Any attribute or method with two leading underscores in its name (but none at the end) is private. It cannot be accessed outside of that class.</li>
       <h4>
           <ul>
               <li> Note : <br> Names with two underscores at the beginning <i>and the end</i> are for built-in methods or attributes for the class</li>
              <li> Note : <br> There is no 'protected' status in Python; so, subclasses would be unable to access these private data either</li>     
           </ul>
       </h4>
    </ul>
</h3>
